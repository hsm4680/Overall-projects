Inner Join : 두 테이블의 중복된 값들만 붙어서 나옴
Outer Join : 중복되어 있지 않은 값들도 붙임. 없는 값들은 NaN으로 들어감.
Left Join : 왼쪽 테이블 기준으로 인덱스가 생김
Right Join : 오른쪽 테이블 기준으로 인덱스가 생김 

```python
df7 = pd.merge(df6,df5, how = 'right')                               #df6를 기준으로 df5를 붙인다. right join
```

```python
df7.to_excel('total_data.xlsx')                 # Excel 파일로 저장해줌
```

```python
df7['날짜(전처리)'] = pd.to_datetime(df7['날짜'])
```

```python
df7['공급월'] = df7['날짜(전처리)'].dt.month
```

```python
pd.pivot_table(data=df7, index = '공급월', values = '재고량', aggfunc='sum').to_excel('result1.xlsx')    #피벗 테이블을 만드는데, 데이터는 df7이고 인덱스는 공급월이고 재고량 값의 합을 값으로 넣는다.
```
